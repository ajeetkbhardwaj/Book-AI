Data Preprocessing using PySpark
1. Load and Read Data

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('data_processing').getOrCreate()
```

```python
# create spark data frame with values from our sample.csv file
df = spark.read.csv('sample.csv', inferSchema=True, header=True)

# operations on the spark data frame
print(df.columns)
print(f"length of features : {len(df.columns)}")
print(f"Total number of records in data frame : {df.count}")

# there is no function that can give shape of data but we can 
print((df.count), (len(df.columns)))

```
